```python
import seaborn as sns
import pandas as pd
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
from pca import pca
import seaborn as sns
plt.rcParams['font.family'] = 'AppleGothic' # For Mac
```

```python
data = pd.read_csv('/Users/csb/Desktop/Project/넙치 머신러닝/0_양식어류 질병예측_221207/20-21-22 통합본.csv',sep=',') # 파일 불러오기
```

```python
plt.rcParams['axes.unicode_minus'] = False
for i in range(len(data.columns)):
    plt.cla()
    ax = sns.stripplot(data=data, x='결과', y=data.columns[i],hue='결과', order=['정상','주의','질병'])
    plt.legend(bbox_to_anchor=(1.2, 1)) # 범례 그래프 밖에서 그리기
    plt.tight_layout() # 잘린 사진 제대로 보이게 하기
    plt.savefig('/Users/csb/Desktop/그래프/{}.jpg'.format(data.columns[i])) # 저장
```
