```python
import pandas as pd
import matplotlib.pyplot as plt
import sys
assert sys.version_info >= (3, 5)
import sklearn
assert sklearn.__version__ >= "0.20"
import numpy as np
import seaborn as sns
from sklearn.model_selection import StratifiedShuffleSplit
import warnings
warnings.filterwarnings(action='ignore')
plt.rcParams['font.family'] = 'AppleGothic' # For Mac
```

```python
data = pd.read_csv("/Users/csb/Desktop/Project/넙치 머신러닝/0_양식어류 질병예측_221207/20,21.csv")
```

```python
data = data.dropna(axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score # 정확도 함수
from sklearn.model_selection import LeaveOneOut
from sklearn.model_selection import cross_val_predict
```

```python
features, labels = data.drop('결과',axis=1), data['결과']
```

```python
forest = RandomForestClassifier(random_state=5, max_features=50, n_estimators=100)

loocv = LeaveOneOut()
score = cross_val_predict(forest,
                        features,
                        labels,
                        cv = loocv)

print("\n## 교차 검증 횟수 : ", len(score))
print('\n## 평균 검증 정확도:', round(np.mean(score),4))
```

```python
from sklearn.metrics import roc_curve, auc
roc_auc_score(labels, score,multi_class='ovr')
```
