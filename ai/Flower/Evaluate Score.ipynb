```python
import pandas as pd
```

```python
data = pd.read_csv('/Users/csb/Desktop/flower matrix.csv', index_col=0)
```

```python
data
```

```python
print("Positive (Jindal)")
print("Negative (Chuljjuk)")
print(" ")
data["Accuracy"]=""
data["Sensitivity"]=""
data["Specificity"]=""
data["Precision"]=""
data["F1"]=""

for i in range(len(data)):
    Accuracy = data.iloc[i,0:2].sum()/data.iloc[i,0:4].sum()
    TP = data['j_to_j'].values[i]/(data['j_to_j'].values[i] + data['j_to_c'].values[i])

    TN = data['c_to_c'].values[i]/(data['c_to_c'].values[i] + data['c_to_j'].values[i])

    FPR = data['c_to_j'].values[i]/(data['c_to_c'].values[i] + data['c_to_j'].values[i])

    Precision = data['j_to_j'].values[i]/(data['j_to_j'].values[i]+data['c_to_j'].values[i])

    F1 = 2/((1/Precision)+(1/TP))
    print("--------- Iter "+str(i+1)+" ---------")
    print("Accuracy : " +str(Accuracy))
    data["Accuracy"].iloc[i] = Accuracy
    print("True Positive Rate(Sensitivity,Recall) : "+str(TP))
    data["Sensitivity"].iloc[i] = TP
    print("True Negative Rate (Specificity) : "+str(TN))
    data["Specificity"].iloc[i] = TN
    print("False Positive Rate : "+str(FPR))
    print("Precision : " +str(Precision))
    data["Precision"].iloc[i] = Precision
    print("F1 : " +str(F1))
    data["F1"].iloc[i] = F1
```

```python
data.iloc[:,4:].to_csv("matrix.csv",header=True)
```
