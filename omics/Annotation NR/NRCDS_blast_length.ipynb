```python
import pandas as pd
```

```python
# Annotation transcript 정보가 들어있는 파일을 열어서 서열간 엔터키 삭제 
f = open("/Users/csb/Desktop/NRCDS_blast.fasta.txt", 'r')
lines = f.readlines()
ff = open("/Users/csb/Desktop/test.txt", "w")

for line in lines:
    if line[0] == '>':
        ff.write("\n"+line)
    else:
        ff.write(line.replace("\n",""))
```

```python
f2 = pd.read_csv("/Users/csb/Desktop/test.txt", sep=' ', header=None)
```

```python
# 각 행의 글자 수 세기
f2['word count'] = f2[0].str.count('')-1
```

```python
f2
```

```python
# 센 글자수를 한칸씩 위로 당겨주기
for i in range(len(f2['word count'])-1):
    f2['word count'][i] = f2['word count'][i+1]
```

```python
# 짝수행(트리니티 ID, 길이)가 들어있는 행 추출
f3 = f2.iloc[::2,:]
```

```python
f3
# 길이 추출 완료
```

```python
# csv 저장
f3.to_csv("NRCDS_blast_length.txt", header=False, sep=' ', index=False)
```
