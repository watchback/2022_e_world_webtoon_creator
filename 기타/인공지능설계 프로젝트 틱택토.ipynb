```python
##문제 5번

### 객체지향 / 인공지능 설계 프로젝트 HW1
### 학번 : 20174373
### 이름 : 김한진
### 제출일자 : 21-09-26

import random

def board_visualization(board):
    ### 틱택토 게임 출력 함수  
    for i_idx in range(3):
        for j_idx in range(3):
            
            if board[i_idx][j_idx] == 0: # not occupied
                print("▨", end='')
            elif board[i_idx][j_idx] == 1: # user
                print("○", end='')
            elif board[i_idx][j_idx] == 2: # computer
                print("×", end='')
            else:
                pass
        print()        
    
def board_turn(board, row, col, val):
    
    if 0 <= row < 3 and 0 <= col < 3:
        if board[row][col] == 0:
            board[row][col] = val
            return 1
        else:
            print("그곳에 돌을 둘 수 없습니다. 다시 입력하세요.")
            return 0
    else:
        print("그곳에 돌을 둘 수 없습니다. 다시 입력하세요.")
        return 0
        
    
def win_condition(v1,v2,v3):  
    if v1 == v2 == v3:  
        if v1 == 1: return 1
        elif v2 == 2: return 2
        else: return 0
    else:
        return 0
        
    
def game_result(board):
    result = 0
    # check rows
    for i_idx in range(3):
        result = win_condition(board[i_idx][0],board[i_idx][1],board[i_idx][2])      
        if result != 0: return result
      
    # check cols
    for i_idx in range(3):
        result = win_condition(board[0][i_idx],board[1][i_idx],board[2][i_idx])
        if result != 0: return result
        
    #check diagonals
    result = win_condition(board[0][0],board[1][1],board[2][2])      
    if result != 0: return result
    result = win_condition(board[0][2],board[1][1],board[2][0])      
    if result != 0: return result
        
    return result

def position():
    row = int(input("돌을 놓을 행을 입력하세요 (0~2):"))
    col = int(input("돌을 놓을 열을 입력하세요 (0~2):"))
    
    return row, col
    
    
    
def computer(board):
    
    row = 0
    col = 0
    empty_cnt = 0
    
    if board[1][1] == 1 and board[0][0] == 0: ## 사용자가 처음에 가운데 두었을 때 컴퓨터가 둘 곳
        row = 0
        col = 0
    elif board[1][1] == 0: ## 사용자가 처음에 가운데 두지 않았을 때 컴퓨터는 가운데에 둔다.
        row = 1
        col = 1
        
    elif board [1][1] == 1 and board[0][0] == 2: ## 사용자가 가운데 두었을 때 두번째 수 부터 알고리즘
        if board[2][0] == 0: ## 모서리가 비어있으면 모서리 공략하는 알고리즘
            row = 2
            col = 0
        elif board[0][2] == 0:
            row = 0
            col = 2
        elif board[2][2] == 0:
            row = 2
            col = 2
        for i_idx in range(3): ## 사용자가 하는 일자공격 방어하는 알고리즘
            if board[0][i_idx] == board[1][i_idx] == 1 and board[2][i_idx] == 0: 
                row = 2
                col = i_idx
            elif board[1][i_idx] == board[2][i_idx] == 1 and board[0][i_idx] == 0:
                row = 0
                col = i_idx
            elif board[0][i_idx] == board[2][i_idx] == 1 and board[1][i_idx] == 0:
                row = 1
                col = i_idx
            elif board[i_idx][0] == board[i_idx][1] == 1 and board[i_idx][2] == 0:
                row = i_idx
                col = 2
            elif board[i_idx][1] == board[i_idx][2] == 1 and board[i_idx][0] == 0:
                row = i_idx
                col = 0
            elif board[i_idx][0] == board[i_idx][2] == 1 and board[i_idx][1] == 0:
                row = i_idx
                col = 1
                    
        if board[2][0] == board[1][1] == 1 : ## 사용자가 하는 대각선 공격 방어하는 알고리즘
            row = 0
            col = 2
        elif board[0][2] == board[1][1] == 1 :
            row = 2
            col = 0
            
        if board[2][0] == board[1][1] == 2 and board[0][2] == 0: ## 컴퓨터의 대각선 공격이 2개 이상 붙어 있을 때 공격 알고리즘
            row = 0
            col = 2
        elif board[0][2] == board[1][1] == 2 and board[2][0] == 0:
            row = 2
            col = 0
        elif board[0][0] == board[1][1] == 2 and board[2][2] == 0:
            row = 2
            col = 2
        elif board[2][2] == board[1][1] == 2 and board[0][0] == 0:
            row = 0
            col = 0
        for i_idx in range(3): ## 컴퓨터의 공격이 2개 이상 붙어 있을 때 컴퓨터에게 차례가 넘어온다면 공격하는 알고리즘
            if board[0][i_idx] == board[1][i_idx] == 2 and board[2][i_idx] == 0: 
                row = 2
                col = i_idx
            elif board[1][i_idx] == board[2][i_idx] == 2 and board[0][i_idx] == 0:
                row = 0
                col = i_idx
            elif board[0][i_idx] == board[2][i_idx] == 2 and board[1][i_idx] == 0:
                row = 1
                col = i_idx
            elif board[i_idx][0] == board[i_idx][1] == 2 and board[i_idx][2] == 0:
                row = i_idx
                col = 2
            elif board[i_idx][1] == board[i_idx][2] == 2 and board[i_idx][0] == 0:
                row = i_idx
                col = 0
            elif board[i_idx][0] == board[i_idx][2] == 2 and board[i_idx][1] == 0:
                row = i_idx
                col = 1
        
            
    elif board[1][1] == 2: ## 사용자가 가운데 두지 않았을 때 두번째 수부터 알고리즘
        if board[2][0] == 0: ## 모서리가 비어있으면 모서리 공략하는 알고리즘
            row = 2
            col = 0
        elif board[0][2] == 0:
            row = 0
            col = 2
        elif board[2][2] == 0:
            row = 2
            col = 2
        elif board[0][0] == 0:
            row = 0
            col = 0
        for i_idx in range(3): ## 사용자가 일자 공격 하였을 때 방어하는 알고리즘
            if board[0][i_idx] == board[1][i_idx] == 1 and board[2][i_idx] == 0: 
                row = 2
                col = i_idx
            elif board[1][i_idx] == board[2][i_idx] == 1 and board[0][i_idx] == 0:
                row = 0
                col = i_idx
            elif board[0][i_idx] == board[2][i_idx] == 1 and board[1][i_idx] == 0:
                row = 1
                col = i_idx
            elif board[i_idx][0] == board[i_idx][1] == 1 and board[i_idx][2] == 0:
                row = i_idx
                col = 2
            elif board[i_idx][1] == board[i_idx][2] == 1 and board[i_idx][0] == 0:
                row = i_idx
                col = 0
            elif board[i_idx][0] == board[i_idx][2] == 1 and board[i_idx][1] == 0:
                row = i_idx
                col = 1
                
            for i_idx in range(3): ## 컴퓨터의 공격이 2개 이상 붙어 있을 때 컴퓨터에게 차례가 넘어온다면 공격하는 알고리즘
                if board[0][i_idx] == board[1][i_idx] == 2 and board[2][i_idx] == 0: 
                    row = 2
                    col = i_idx
                elif board[1][i_idx] == board[2][i_idx] == 2 and board[0][i_idx] == 0:
                    row = 0
                    col = i_idx
                elif board[0][i_idx] == board[2][i_idx] == 2 and board[1][i_idx] == 0:
                    row = 1
                    col = i_idx
                elif board[i_idx][0] == board[i_idx][1] == 2 and board[i_idx][2] == 0:
                    row = i_idx
                    col = 2
                elif board[i_idx][1] == board[i_idx][2] == 2 and board[i_idx][0] == 0:
                    row = i_idx
                    col = 0
                elif board[i_idx][0] == board[i_idx][2] == 2 and board[i_idx][1] == 0:
                    row = i_idx
                    col = 1
                
            if board[2][0] == board[1][1] == 2 and board[0][2] == 0: ## 컴퓨터의 대각선 공격이 2개 이상 붙어 있을 때 공격 알고리즘
                row = 0
                col = 2
            elif board[0][2] == board[1][1] == 2 and board[2][0] == 0:
                row = 2
                col = 0
            elif board[0][0] == board[1][1] == 2 and board[2][2] == 0:
                row = 2
                col = 2
            elif board[2][2] == board[1][1] == 2 and board[0][0] == 0:
                row = 0
                col = 0
    
    else: 
        for i_idx in range(3): ## 컴퓨터의 공격이 2개 이상 붙어 있을 때 컴퓨터에게 차례가 넘어온다면 공격하는 알고리즘
            if board[0][i_idx] == board[1][i_idx] == 2 and board[2][i_idx] == 0: 
                row = 2
                col = i_idx
            elif board[1][i_idx] == board[2][i_idx] == 2 and board[0][i_idx] == 0:
                row = 0
                col = i_idx
            elif board[0][i_idx] == board[2][i_idx] == 2 and board[1][i_idx] == 0:
                row = 1
                col = i_idx
            elif board[i_idx][0] == board[i_idx][1] == 2 and board[i_idx][2] == 0:
                row = i_idx
                col = 2
            elif board[i_idx][1] == board[i_idx][2] == 2 and board[i_idx][0] == 0:
                row = i_idx
                col = 0
            elif board[i_idx][0] == board[i_idx][2] == 2 and board[i_idx][1] == 0:
                row = i_idx
                col = 1
        if board[2][0] == board[1][1] == 2 and board[0][2] == 0: ## 컴퓨터의 대각선 공격이 2개 이상 붙어 있을 때 공격 알고리즘
            row = 0
            col = 2
        elif board[0][2] == board[1][1] == 2 and board[2][0] == 0:
            row = 2
            col = 0
        elif board[0][0] == board[1][1] == 2 and board[2][2] == 0:
            row = 2
            col = 0
        elif board[2][2] == board[1][1] == 2 and board[0][0] == 0:
            row = 0
            col = 0
        elif board[2][0] == 0: ## 모서리가 비어있으면 모서리 공략하는 알고리즘
            row = 2
            col = 0
        elif board[0][2] == 0:
            row = 0
            col = 2
        elif board[2][2] == 0:
            row = 2
            col = 2
        elif board[0][0] == 0:
            row = 0
            col = 0
                
        else: ## 위의 모든 경우가 없으면 랜덤으로 둔다.
            for i_idx in range(3):
                for j_idx in range(3):
                    if board[i_idx][j_idx] == 0:
                        empty_cnt+=1

            rnd_idx = random.randrange(1,empty_cnt)
            empty_cnt = 0
    
            for i_idx in range(3):
                for j_idx in range(3):
                    if board[i_idx][j_idx] == 0:          
                        if empty_cnt == rnd_idx:
                            row = i_idx
                            col = j_idx
                            return row, col
                    
                        empty_cnt+=1
        
    return row, col


### 게임 시작
print("게임 시작")
board = [[0,0,0],[0,0,0],[0,0,0]]
board_visualization(board)
victory = 0

for i_idx in range(5): # 플레이어는 최대 5번 돌을 둔다
    ###### User 공격 ######
    print("============= Your turn =============")
    row, col = position()
    while board_turn(board, row, col,1) == 0:
        row, col = position()
        
    board_visualization(board)
    if game_result(board) == 1:
        victory = 1
        print("사용자가 승리하였습니다.")
        break
    elif game_result(board) == 2:
        victory = 2
        print("컴퓨터가 승리하였습니다.")
        break      
    ########################
    
    
    ###### 컴퓨터 공격 ######
    print("============= Computer =============")
    row, col = computer(board)
    board_turn(board, row, col, 2)
    board_visualization(board)   
    if game_result(board) == 1:
        victory = 1
        print("사용자가 승리하였습니다.")
        break
    elif game_result(board) == 2:
        victory = 2
        print("컴퓨터가 승리하였습니다.")
        break
    ########################

if victory == 0:
    print("비겼습니다.")
```
