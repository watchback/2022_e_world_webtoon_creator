```python
import seaborn as sns
import pandas as pd
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
from pca import pca
import seaborn as sns
```

```python
r_all = pd.read_csv('/Users/csb/R_all.csv',sep=',',index_col=0) # 파일 불러오기
r_all
r_log = pd.read_csv('/Users/csb/R_log.csv',sep=',',index_col=0) # 파일 불러오기
r_log
r_cutoff = pd.read_csv('/Users/csb/R_cutoff.csv',sep=',',index_col=0) # 파일 불러오기
r_cutoff
r_log_cutoff = pd.read_csv('/Users/csb/R_log_cutoff.csv',sep=',',index_col=0) # 파일 불러오기
r_log_cutoff
```

```python
import seaborn as sns
fig, ax = plt.subplots(figsize=(7,7))
sns.scatterplot(data=r_all,x='PC1',y='PC2',hue=r_all['category'], s=200) # absolute, All genes
plt.xlabel('PC1 (0.9155%)', fontsize=15)
plt.ylabel('PC2 (0.08213%)', fontsize=15)
for line in range(0,r_all.shape[0]):
     plt.text(r_all.PC1[line]+0.0005, r_all.PC2[line], r_all.index[line], horizontalalignment='left', size='medium', color='black', weight='semibold')
plt.savefig('/Users/csb/Desktop/all.jpg')
plt.show()
```
